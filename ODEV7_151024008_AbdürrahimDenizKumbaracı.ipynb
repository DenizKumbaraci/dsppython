```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
from scipy.io import loadmat
import scipy.signal as sgnl


```

### SORU-1

```python
filter_data = loadmat('FIRbandpassGauss.mat')
Coeffs = filter_data['ba'].astype(np.float)
b = Coeffs[0,:]

w,Hw = sgnl.freqz(b,1)

Hw_mag=np.abs(Hw)
Hw_phs=np.unwrap(np.angle(Hw))
Hw_gdly=-np.diff(Hw_phs)


plt.figure()

plt.subplot(3,1,1)
plt.title('Characteristics of FIR Low-Pass Filter(Gaussian)')
plt.plot(w/np.pi ,Hw_mag )
plt.ylabel('|H(e^jw)|')
plt.subplot(3,1,2)
plt.plot(w/np.pi ,Hw_phs)
plt.ylabel('Phase H(e^jw)')
plt.subplot(3,1,3)
plt.plot(w[1::]/np.pi ,Hw_gdly )
plt.ylabel('Group Delay H(e^jw)')
plt.xlabel('Normalised Frequency (x pi rad/sample)')
```

## İki filtre için de geçirme frekanslarında faz bileşeninin doğrusal olduğunu ve buna bağlı olarak grup gecikmelerinin sabit değer olduğunu grafiklerde görebilyoruz. Bu sebeple bu filtrelerin uygulanan şaretleri bozulmasız şekilde filtreliyeceğini söyleyebiliriz. Blackman metodunun kesim frekansı etrafında gauss methoduna göre daha ideale yakın bir frekans cevabı verdiğini fakat gauss metodlu filtrenin kesim frekansı dışında da doğrusal faz grafiği olduğunu gözlemleyebiliriz(geçirme bantlarında negatif değer almadığı için nedenseldir).

```python
filter_data = loadmat('IIRLowpassCheby1.mat')
Coeffs = filter_data['ba'].astype(np.float)
b = Coeffs[0,:]
a = Coeffs[1,:]
w,Hw = sgnl.freqz(b,a)

Hw_mag=np.abs(Hw)
Hw_phs=np.unwrap(np.angle(Hw))
Hw_gdly=-np.diff(Hw_phs)


plt.figure()

plt.subplot(3,1,1)
plt.title('Characteristics of IIR Low-Pass Filter(Chebvyshev 1)')
plt.plot(w/np.pi ,Hw_mag )
plt.ylabel('|H(e^jw)|')
plt.subplot(3,1,2)
plt.plot(w/np.pi ,Hw_phs)
plt.ylabel('Phase H(e^jw)')
plt.subplot(3,1,3)
plt.plot(w[1::]/np.pi ,Hw_gdly )
plt.ylabel('Group Delay H(e^jw)')
plt.xlabel('Normalised Frequency (x pi rad/sample)')
```

```python
filter_data = loadmat('IIRHighpassButterworth.mat')
Coeffs = filter_data['ba'].astype(np.float)
b = Coeffs[0,:]
a = Coeffs[1,:]
w,Hw = sgnl.freqz(b,a)

Hw_mag=np.abs(Hw)
Hw_phs=np.unwrap(np.angle(Hw))
Hw_gdly=-np.diff(Hw_phs)


plt.figure()

plt.subplot(3,1,1)
plt.title('Characteristics of IIR High-Pass Filter(Butterworth)')
plt.plot(w/np.pi ,Hw_mag )
plt.ylabel('|H(e^jw)|')
plt.subplot(3,1,2)
plt.plot(w/np.pi ,Hw_phs)
plt.ylabel('Phase H(e^jw)')
plt.subplot(3,1,3)
plt.plot(w[1::]/np.pi ,Hw_gdly )
plt.ylabel('Group Delay H(e^jw)')
plt.xlabel('Normalised Frequency (x pi rad/sample)')
```
