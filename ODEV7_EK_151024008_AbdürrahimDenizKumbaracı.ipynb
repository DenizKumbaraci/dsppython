```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
from scipy.io import loadmat
import scipy.signal as sgnl


```

## Ön hazırlık ödevleri (ÖDEV-1 başlığı altındaki soruların çözümleri aşağıda verilecek) 

### Type I

```python
N=100
hn1=np.array([1,1,1,1,1,1])
w=np.arange(-np.pi,np.pi,2*np.pi/N)
Hn1=np.fft.fftshift(np.fft.fft(hn1,N))

plt.figure()
plt.subplot(3,1,1)
plt.title('Characteristics of Type I Filter')
plt.plot(w/np.pi,np.abs(Hn1))
plt.ylabel('|H(e^jw)|')
plt.subplot(3,1,2)
plt.plot(w/np.pi,np.unwrap(np.angle(Hn1)))
plt.ylabel('Phase H(e^jw)')
plt.subplot(3,1,3)
plt.plot(w[1::]/np.pi,-np.diff(np.angle(Hn1)))
plt.xlabel('Normalised Frequency (x pi rad/sample)')
plt.ylabel('Group Delay H(e^jw)')
```

## Type II

```python
hn2=np.array([1,1,1,1,1,1,1])
w=np.arange(-np.pi,np.pi,2*np.pi/N)
Hn2=np.fft.fftshift(np.fft.fft(hn2,N))

plt.figure()
plt.subplot(3,1,1)
plt.title('Characteristics of Type II Filter')
plt.plot(w/np.pi,np.abs(Hn2))
plt.ylabel('|H(e^jw)|')
plt.subplot(3,1,2)
plt.plot(w/np.pi,np.unwrap(np.angle(Hn2)))
plt.ylabel('Phase H(e^jw)')
plt.subplot(3,1,3)
plt.plot(w[1::]/np.pi,-np.diff(np.angle(Hn2)))
plt.ylabel('Group Delay H(e^jw)')
plt.xlabel('Normalised Frequency (x pi rad/sample)')
```

## Type III

```python
hn3=np.array([1,1,1,-1,-1,-1])
w=np.arange(-np.pi,np.pi,2*np.pi/N)
Hn3=np.fft.fftshift(np.fft.fft(hn3,N))

plt.figure()
plt.subplot(3,1,1)
plt.title('Characteristics of Type III Filter')
plt.plot(w/np.pi,np.abs(Hn3))
plt.ylabel('|H(e^jw)|')
plt.subplot(3,1,2)
plt.plot(w/np.pi,np.unwrap(np.angle(Hn3)))
plt.ylabel('Phase H(e^jw)')
plt.subplot(3,1,3)
plt.plot(w[1::]/np.pi,-np.diff(np.angle(Hn3)))
plt.xlabel('Normalised Frequency (x pi rad/sample)')
plt.ylabel('Group Delay H(e^jw)')
```

## Type IV

```python
hn4=np.array([1,1,1,0,-1,-1,-1])
w=np.arange(-np.pi,np.pi,2*np.pi/N)
Hn4=np.fft.fftshift(np.fft.fft(hn4,N))

plt.figure()
plt.subplot(3,1,1)
plt.title('Characteristics of Type IV Filter')
plt.plot(w/np.pi,np.abs(Hn4))
plt.ylabel('|H(e^jw)|')
plt.subplot(3,1,2)
plt.plot(w/np.pi,np.unwrap(np.angle(Hn4)))
plt.ylabel('Phase H(e^jw)')
plt.subplot(3,1,3)
plt.plot(w[1::]/np.pi,-np.diff(np.angle(Hn4)))
plt.xlabel('Normalised Frequency (x pi rad/sample)')
plt.ylabel('Group Delay H(e^jw)')
```

## Grafikler incelendiği zaman tip 1 ve 2 filtrelerin alçak geçiren filtre olduklarını ve geçirme bantlarında grup gecikmelerinin sabit olduğunu, tip 3 ve 4 filtrelerin bant geçiren filtreler olduğunu ve benzer şekilde grup gecikmelerinin geçirme bantlarında sabit olduğunu görebilmekteyiz.

```python
n=np.arange(0,N)
xn= np.cos(0.2*np.pi*n) + np.cos(0.4*np.pi*n - np.pi/2)  + np.cos(0.8*np.pi*n - np.pi/5)
Xn= np.fft.fftshift(np.fft.fft(xn,N))


Yn1=Xn*Hn1
Yn2=Xn*Hn2
Yn3=Xn*Hn3
Yn4=Xn*Hn4


yn1=np.fft.ifft(np.fft.ifftshift(Yn1))
yn2=np.fft.ifft(np.fft.ifftshift(Yn2))
yn3=np.fft.ifft(np.fft.ifftshift(Yn3))
yn4=np.fft.ifft(np.fft.ifftshift(Yn4))



plt.figure()
plt.title('Type I Filter I/O')
plt.subplot(2,2,1)

plt.stem(n,xn)
plt.title('x[n]')
plt.subplot(2,2,2)
plt.stem(w/np.pi,np.abs(Xn))
plt.title('|X(e^jw)|')
plt.subplot(2,2,3)
plt.stem(n,yn1)
plt.title('y1[n]')
plt.xlabel('n(samples)')
plt.subplot(2,2,4)
plt.stem(w/np.pi,np.abs(Yn1))
plt.title('|Y1(e^jw)|')
plt.xlabel('Normalised Frequency (x pi rad/sample)')



plt.figure()
plt.title('Type II Filter I/O')
plt.subplot(2,2,1)

plt.stem(n,xn)
plt.title('x[n]')
plt.subplot(2,2,2)
plt.stem(w/np.pi,np.abs(Xn))
plt.title('|X(e^jw)|')
plt.subplot(2,2,3)
plt.stem(n,yn2)
plt.title('y2[n]')
plt.xlabel('n(samples)')
plt.subplot(2,2,4)
plt.stem(w/np.pi,np.abs(Yn2))
plt.title('|Y2(e^jw)|')
plt.xlabel('Normalised Frequency (x pi rad/sample)')



plt.figure()
plt.title('Type III Filter I/O')
plt.subplot(2,2,1)

plt.stem(n,xn)
plt.title('x[n]')
plt.subplot(2,2,2)
plt.stem(w/np.pi,np.abs(Xn))
plt.title('|X(e^jw)|')
plt.subplot(2,2,3)
plt.stem(n,yn3)
plt.title('y3[n]')
plt.xlabel('n(samples)')
plt.subplot(2,2,4)
plt.stem(w/np.pi,np.abs(Yn3))
plt.title('|Y3(e^jw)|')
plt.xlabel('Normalised Frequency (x pi rad/sample)')




plt.figure()
plt.title('Type IV Filter I/O')
plt.subplot(2,2,1)

plt.stem(n,xn)
plt.title('x[n]')
plt.subplot(2,2,2)
plt.stem(w/np.pi,np.abs(Xn))
plt.title('|X(e^jw)|')
plt.subplot(2,2,3)
plt.stem(n,yn4)
plt.title('y4[n]')
plt.xlabel('n(samples)')
plt.subplot(2,2,4)
plt.stem(w/np.pi,np.abs(Yn4))
plt.title('|Y4(e^jw)|')
plt.xlabel('Normalised Frequency (x pi rad/sample)')
```

## Tip 3 ve 4 filtreler bant geçiren filtreler olsa da giriş işaretin düşük frekanslı bileşenleri geçirme bandında, yüksek frekanslı bileşenler bandın dışında kaldığı için bütün filtreler alçak geçiren filtre gibi davranmıştır. Çıkış işaretinin frekans cevabını incelediğimizde genliklerinin filtrelerin zarfına göre sönümlenip artırılmış olduğunu görebilmekteyiz. İşaretlerin zaman ifadelerine baktğımızda ise grup gecikmelerinden dolayı tepe noktalarının beklendiği gibi sağ tarafa ötelendiğini görebilmekteyiz(Filtrelerin hepsi nedenseldir).
