## I)

```python
Fc=9600
Fs=8192
Ts=1/Fs
Tsc=2/Fc
N=128
n=np.arange(0,3*N)
t=np.arange(0,3*Tsc,3*Tsc/1000)
x3=np.sin(9600/8192*np.pi*n)
x3c=np.sin(Fc*np.pi*t)

plt.figure()
plt.plot(t,x3c)

plt.figure()
plt.stem(n,x3)





w=np.linspace(-np.pi,np.pi-(2*np.pi/N),N)
x3w=np.fft.fft(x3,N)/N
x3w=np.fft.fftshift(x3w)

plt.figure()
plt.subplot(1,2,1)
plt.stem(w,np.abs(x3w))
plt.title('|X1(w)|')
plt.xlabel('w')
plt.subplot(1,2,2)
plt.stem(w,np.angle(x3w))
plt.title('Phase of X1(w)')
plt.xlabel('w')
```

### SORU-2

```python
def quantize(x, S):
    X = x.reshape((-1,1))
    S = S.reshape((1,-1))
    dists = abs(X-S)
    nearestIndex = dists.argmin(axis=1)
    quantized = S.flat[nearestIndex]
    return quantized.reshape(x.shape)


n3=np.arange(0,9)
xarr=np.array([1.25 , 3.5 , 4 , 6.7 , - 1.4 , - 3.5 , -10.5 , 8.5 , 3.8])
k_bit= 3
Max_range= 19
q=pow(2,k_bit)
delta = Max_range/q
S_midrise = -np.amax(xarr) +delta/2+np.arange(q)*delta
S_midtread = -np.amax(xarr)  +np.arange(q)*delta

y_midrise = quantize(xarr, S_midrise)
y_midtread = quantize(xarr, S_midtread)

plt.figure()
plt.stem(n3,xarr,label='Giriş Dizisi')
plt.stem(n3,y_midrise,'g',label='Mid-Rise ile kuantalanmış işaret',markerfmt='go')
plt.title('Kuantalanmış ve orjinal dizi')
plt.legend()



plt.figure()
plt.stem(n3,xarr,label='Giriş Dizisi')
plt.stem(n3,y_midtread,'g',label='Mid-Tread ile kuantalanmış işaret',markerfmt='go')
plt.title('Kuantalanmış ve orjinal dizi')
plt.legend()
```
