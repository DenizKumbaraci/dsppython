```python
ekcos=np.array([0,0.707106,-1,0.707106,0,-0.707106,1,-0.707106])
fftcos=fft(ekcos,8)
w=np.arange(0,2*np.pi,np.pi/4)
plt.figure()
plt.subplot(121)
plt.stem(w,np.abs(fftcos))
plt.xlabel('Normalize Frekans (x pi)')
plt.title('|A Cos(wn + phi)|')
plt.subplot(122)
plt.stem(np.angle(fftcos))
plt.title('A Cos(wn + phi) nin Açısı')
plt.xlabel('Normalize Frekans (x pi)')

print(np.abs(fftcos))
print(np.angle(fftcos))
```

Elde edilen sonuçlara bakarak cosinusun genliğinin 8 ve fazının frekansa göre değiştiğini görüyoruz. Fakat giriş verilerine bakınca sabit bir faz ve 1 genlik değeri beklerdik.

```python
n=np.arange(0,11)
w_var=n*np.pi/3
H_ejw=(1-np.exp(-1j*w_var))

magH=abs(H_ejw)
phsH=np.angle(H_ejw)

yn=magH*np.cos(np.pi/3 + phsH)

plt.figure()
plt.stem(n,yn)
plt.title('y[n]')
plt.xlabel('n(samples)')
```
