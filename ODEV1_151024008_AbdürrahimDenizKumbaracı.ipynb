Buna göre işareti iki periyor boyunca çizdirelim.(Burada dizini oluşturmak için arange komutunu kullanırsak n=20 noktası dahil olmayacaktır böylelikle tam 2 periyod boyunca sinyali çizdirmiş olacağız.) 

```python
n=np.arange(0,20,1)
x=(2*np.sin((2*np.pi*n/5)-(np.pi/5)))+(np.sin((3*np.pi*n/5)-(2*np.pi/5)))
plt.title('x[n]')
plt.stem(n,x)
plt.show()

```

```python
theta=np.pi/5

n1=np.arange(0,16)
s=np.exp(1j*theta)
x2=np.linspace(s,s,16)
plt.title('Real Part Of x2[n]')
plt.stem(n1,x2.real)
fig=plt.figure()
plt.title('Imaginary Part Of x2[n]')
plt.stem(n1,x2.imag)
n2=np.arange(-10,5,)
x6=np.cos(np.pi*n2/5)+1j*np.sin(np.pi*n2/5)
fig=plt.figure()
plt.title('Imaginary Part Of x6[n]')
plt.stem(n2,x6.imag)
fig=plt.figure()
plt.title('Real Part Of x6[n]')
plt.stem(n2,x6.real)
```
