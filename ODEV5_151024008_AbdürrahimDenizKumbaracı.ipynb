### SORU-1

fonksiyonlarda t=nTs yazarsak;

x1[n]=7cos⁡(𝜋n/11) , x2[n]= 3cos(𝜋n/11)+4cos⁡(9𝜋n/11)


2𝜋k/N = 9𝜋/11 =>  k=9 , N=22


```python
Fs=2200
Ts=1/Fs
N=22
n=np.arange(0,2*N)
x1=7*np.cos(np.pi*n/11)
x2=3*np.cos(np.pi*n/11) + 4*np.cos(9/11*np.pi*n)


plt.figure()
plt.stem(n,x1)
plt.xlabel('n(örnek)')
plt.title('x1[n]')


plt.figure()
plt.stem(n,x2)
plt.xlabel('n(örnek)')
plt.title('x2[n]')



```

x2 işaretinin yüksek frekanslı bileşeni için örnekleme frekansı nyquist frekansının altında kalmış olsa bile 2pi ile tümleyen olmadığı için x1 ve x2 farklı çıkmıştır.

```python
# w=np.arange(-np.pi,np.pi,2*np.pi/N) 
w=np.linspace(-np.pi,np.pi-(2*np.pi/N),N)
x1w=np.fft.fft(x1,N)/N
x1w=np.fft.fftshift(x1w)


# w=np.arange(-np.pi,np.pi,2*np.pi/N) 
w=np.linspace(-np.pi,np.pi-(2*np.pi/N),N)
x2w=np.fft.fft(x2,N)/N
x2w=np.fft.fftshift(x2w)

plt.figure()
plt.subplot(1,2,1)
plt.stem(w,np.abs(x1w))
plt.title('|X1(w)|')
plt.xlabel('w')
plt.subplot(1,2,2)
plt.stem(w,np.angle(x1w))
plt.title('Phase of X1(w)')
plt.xlabel('w')



plt.figure()
plt.subplot(1,2,1)
plt.stem(w,np.abs(x2w))
plt.title('|X2(w)|')
plt.xlabel('w')
plt.subplot(1,2,2)
plt.stem(w,np.angle(x2w))
plt.title('Phase of X2(w)')
plt.xlabel('w')
```

Görüldüğü üzere X1(w) işareti ve X2(w) işaretleri beklendiği gibi farklı çıkmıştır (ikinci işaretin bileşenleri bu örnekleme hızında birbirlerini 2𝜋 ye tamamlamadıkları için) fakat yine de x2 işareti nyquist hızının altında örneklenmiştir.

X1(w) 2𝜋f = 0.285 yani yaklaşık +-𝜋/11 de genliği 3.5 olan iki bileşenden oluşmaktadır. Yani bir sinusoid olduğunu görebilmekteyiz. Faz grafiğini hesaba katmazsak 

x1[n]= 2*(3.5)cos(𝜋n/11 + phi) şeklinde bir işaret olduğunu tahmin edebiliriz.

faz grafiğine baktığımızda bu işareti 

x1[n]=7cos(𝜋n/11) olarak geri elde edebiliriz

benzer bir şekilde ikinci grafiğin 𝜋/11 ve 9𝜋/11 frekanslarında dinusoidler olduğunu görebiliriz.

x2[n]= 3cos(𝜋n/11) + 4cos(9𝜋n/11) olarak elde ederiz.
