```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
a=3-4j
b=1+2j
C=a+b
print("C sayısının Gerçek kısmı:")
print(C.real)
print("C sayısının Sanal kısmı:")
print(C.imag)
print("C sayısının Genlik değeri:")
print(np.abs(C))
print("C sayısının Faz değeri:")
print(np.angle(C))
```

### SORU-2

İlk işaret için ödev kısmında da bulduğumuz gibi periyot 5, ikinci işaret için 10 dur.Sırasıyla kodlarını yazarsak;

```python

n1=np.arange(0,10)
x1=(2*np.sin((2*np.pi*n1/5)-(np.pi/5)))
plt.title('x1[n]')
plt.stem(n1,x1)
plt.show()


n2=np.arange(0,20)
x2=np.sin((3*np.pi*n2/5)-(2*np.pi/5))
fig=plt.figure()
plt.title('x2[n]')
plt.stem(n2,x2)
plt.show()
```

### Soru - 3

```python
theta=np.pi/5
n=np.arange(0,16)
s=np.exp(theta)
x1=np.linspace(s,s,16)
x3=np.exp(1j*theta*n)
s2=np.cos(np.pi/5)+1j*np.sin(np.pi/5)
x4=np.linspace(s2,s2,16)
x5=np.cos(np.pi*n/5)+1j*np.sin(np.pi*n/5)


plt.title('x1[n]')
plt.stem(n,x1)

fig=plt.figure()

plt.title('Imaginary Part Of x3[n]')
plt.stem(n,x3.imag)
fig=plt.figure()
plt.title('Real Part Of x3[n]')
plt.stem(n,x3.real)

fig=plt.figure()

plt.title('Imaginary Part Of x4[n]')
plt.stem(n,x4.imag)
fig=plt.figure()
plt.title('Real Part Of x4[n]')
plt.stem(n,x4.real)

fig=plt.figure()

plt.title('Imaginary Part Of x5[n]')
plt.stem(n,x5.imag)
fig=plt.figure()
plt.title('Real Part Of x5[n]')
plt.stem(n,x5.real)


```
