```python
h=[0,1,0,-1]
n=np.arange(0,20)
x1u=x1u=(np.heaviside(n,1)-np.heaviside(n-20,1))
x1=np.cos(np.pi*n/5)*x1u

plt.figure()
plt.title('x1[n]')
plt.stem(n,x1)

x2u=(np.heaviside(n-1,1)-np.heaviside(n-5,1))
x2=pow(-0.3,n)*x2u
plt.figure()
plt.title('x2[n]')
plt.stem(n,x2)



plt.figure()
plt.title('y1[n]')
plt.stem(n,np.convolve(x1,h,'same'))


plt.figure()
plt.title('y2[n]')
plt.stem(n,np.convolve(x2,h,'same'))
```
