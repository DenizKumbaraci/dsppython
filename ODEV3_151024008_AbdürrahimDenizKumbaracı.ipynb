### SORU-1

H(e^jw)= n=-sonsuzdan sonsuza toplam(h[n].e^-jwn) dir. Bu denklemde h[n] i yerine koyucak olursak bu toplam, n=-sonsuzdan sonsuza toplam((dirac[n]-dirac[n-1]).e^-jwn) olacaktır.Bu toplam sadece n=0 ve n=1 değerleri için 0'dan farklı değer alıcaktır. Bu n değerlerini yerine koyarsak; 1-e^-jw = H(e^jw) sonucuna ulaşırız.

```python
w_var=np.arange(0,2*np.pi,2*np.pi/1000)
H_ejw=(1-np.exp(-1j*w_var))
magH=abs(H_ejw)
phsH=np.array([-w if w<np.pi else -w+np.pi for w in w_var])
phsH=np.angle(H_ejw)




plt.figure()
plt.subplot(121)
plt.plot(w_var/np.pi,magH)
plt.title('|H(e^jw)|')
plt.xlabel('Normalize Frekans(x pi)')
plt.subplot(122)
plt.plot(w_var/np.pi,phsH)
plt.title('H(e^jw) in Faz Grafiği')
plt.xlabel('Normalize Frekans(x pi)')
```
